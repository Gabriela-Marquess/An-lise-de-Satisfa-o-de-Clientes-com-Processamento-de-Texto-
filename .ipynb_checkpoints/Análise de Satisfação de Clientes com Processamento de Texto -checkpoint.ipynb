```python
# Carregar a lista de stopwords do NLTK
stop_words = set(stopwords.words('portuguese'))
```

```python
# Função para limpar o texto
def limpar_texto(texto):
    # Remover pontuação e transformar em minúsculas
    texto = ''.join([char for char in texto if char.isalnum() or char.isspace()])
    texto = texto.lower()
    texto = ' '.join([palavra for palavra in texto.split() if palavra not in stop_words])
    return texto
```

```python
# Aplicar a função de limpeza na coluna 'Comentário'
df['Comentário Limpo'] = df['Comentário'].apply(limpar_texto)
```

```python
# Exibir as primeiras linhas para verificar
df[['Comentário', 'Comentário Limpo']].head()
```

```python
# Gerar a nuvem de palavras
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(df['Comentário Limpo']))
```

```python
# Exibir a nuvem de palavras
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from textblob import TextBlob
```

```python
# Função para calcular o sentimento (polaridade)
def analisar_sentimento(texto):
    # TextBlob retorna um valor de polaridade entre -1 (negativo) e 1 (positivo)
    return TextBlob(texto).sentiment.polarity
```

```python
# Aplicar a função na coluna 'Comentário Limpo'
df['Sentimento'] = df['Comentário Limpo'].apply(analisar_sentimento)
```

```python
# Adicionar uma coluna de "Classificação" com base na polaridade
df['Classificação Sentimento'] = df['Sentimento'].apply(lambda x: 'Positivo' if x > 0 else ('Negativo' if x < 0 else 'Neutro'))
```

```python
# Exibir as primeiras linhas para verificar
df[['Comentário', 'Sentimento', 'Classificação Sentimento']].head()
```

```python
# Gráfico de distribuição dos sentimentos
plt.figure(figsize=(8, 6))
sns.countplot(x='Classificação Sentimento', data=df, palette='viridis')
plt.title('Distribuição dos Sentimentos dos Comentários')
plt.xlabel('Sentimento')
plt.ylabel('Contagem')
plt.show()
```

```python
# Contagem dos sentimentos
sentimentos_count = df['Classificação Sentimento'].value_counts()
```

```python
# Exibir o resumo dos sentimentos
sentimentos_count
```

```python
# Comentários mais positivos
comentarios_positivos = df[df['Classificação Sentimento'] == 'Positivo'].sort_values(by='Sentimento', ascending=False).head(5)
```

```python
# Comentários mais negativos
comentarios_negativos = df[df['Classificação Sentimento'] == 'Negativo'].sort_values(by='Sentimento', ascending=True).head(5)
```

```python
# Exibir os resultados
comentarios_positivos[['Comentário', 'Sentimento']], comentarios_negativos[['Comentário', 'Sentimento']]
```

```python
# Exibir a média da polaridade dos sentimentos
media_sentimento = df['Sentimento'].mean()
print(f"Média da Polaridade dos Comentários: {media_sentimento:.2f}")
```

```python
# Conclusões gerais sobre a análise
if media_sentimento > 0:
    print("Os comentários são em sua maioria positivos.")
elif media_sentimento < 0:
    print("Os comentários são em sua maioria negativos.")
else:
    print("Os comentários são neutros.")
```
